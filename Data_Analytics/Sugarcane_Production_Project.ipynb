```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("List of Countries by Sugarcane Production.csv")
data
```

```python
data.shape
```

**DATA CLEANING**

```python
data.head()
```

```python
data.isna()
```

```python
data.isnull().sum()
```

```python
data
```

```python
data["Production (Tons)"]=data["Production (Tons)"].str.replace(".","")
data["Production per Person (Kg)"]=data["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
data["Acreage (Hectare)"]=data["Acreage (Hectare)"].str.replace(".","")
data["Yield (Kg / Hectare)"]=data["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")


data.head()
```

```python
data.rename(columns={"Production (Tons)":"Production(Tons)"},inplace=True)
data.rename(columns={"Production per Person (Kg)":"ProductionperPerson(Kg)"},inplace=True)
data.rename(columns={"Acreage (Hectare)":"Acreage(Hectare)"},inplace=True)
data.rename(columns={"Yield (Kg / Hectare)":"Yield(Kg/Hectare)"},inplace=True)
data.head()
```

```python
data.dropna().reset_index()
```

```python
data
```

```python
data.dtypes
```

```python
data["Production(Tons)"]=data["Production(Tons)"].astype(float)
data["ProductionperPerson(Kg)"]=data["ProductionperPerson(Kg)"].astype(float)
data["Acreage(Hectare)"]=data["Acreage(Hectare)"].astype(float)
data["Yield(Kg/Hectare)"]=data["Yield(Kg/Hectare)"].astype(float)

```

```python
data.dtypes
```

```python
data.nunique()
```

**BIVARIATE ANALYSIS**

**Which country produces maximum sugarcane?**

```python
data.head()
```

```python
data_new=data[["Country","Production(Tons)"]].set_index("Country")
data_new.head()
```

```python
data_new["Production(Tons)_percent"]=(data_new["Production(Tons)"]*100)/data_new["Production(Tons)"].sum()
data_new.head()
```

```python

data_new["Production(Tons)"].plot(kind="bar")
```

```python
data_new["Production(Tons)"].head().plot(kind="bar")
```

```python
data_new["Production(Tons)_percent"].plot(kind="pie",autopct="% .2f")
```

```python
data_new["Production(Tons)_percent"].head(10).plot(kind="pie",autopct="% .2f")
```

```python
ax=sns.barplot(data=data_new.head(10), x="Country", y="Production(Tons)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

**Which Country has Highest land?**

```python
data.head()
```

```python
df=data[["Country","Acreage(Hectare)"]].set_index("Acreage(Hectare)")
df.head()
```

```python
df_acr=df.sort_values("Acreage(Hectare)",ascending=False)
ax=sns.barplot(data=df_acr.head(10), x="Country", y="Acreage(Hectare)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

**Which Country Has Highest Yield Per hectare?**

```python
df_yield=data.sort_values("Yield(Kg/Hectare)",ascending=False)
ax=sns.barplot(data=df_yield.head(20), x="Country", y="Yield(Kg/Hectare)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

**Which Country Has Highest Production?**

```python
df_prod=data.sort_values("ProductionperPerson(Kg)",ascending=False)
ax=sns.barplot(data=df_prod.head(10), x="Country", y="ProductionperPerson(Kg)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```
