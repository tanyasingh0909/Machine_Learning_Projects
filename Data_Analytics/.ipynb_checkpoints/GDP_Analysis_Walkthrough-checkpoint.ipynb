*4)Whats the % of GDP growth of India in Year 1994 (*as compare to 1993)?**


```python
a = data.loc[(data["Year"] == 1994) & (data["Country Name"] == "India")]
value1 = a["Value"].values[0]  # Extract the value as a scalar
print(value1)
```

```python

a = data.loc[(data["Year"] == 1993) & (data["Country Name"] == "India")]
value2 = a["Value"].values[0]  # Extract the value as a scalar
print(value2)
```

```python
percentage_gdp_growth=(value1-value2)*100/value1
print(percentage_gdp_growth,"%")
```

```python
gdp_inc=[0]
for i in range(1, len(data)):
    curr_year=data.iloc[i, 3]
    prev_year=data.iloc[i-1,3]
    gdp_inc.append(round((((curr_year-prev_year) *100) /curr_year),2))
    
```

```python
d=data.assign(GDP=gdp_inc)
d
```
